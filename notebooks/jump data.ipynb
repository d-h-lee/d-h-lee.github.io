```python
import pandas as pd
```

```python
file_name = '1단 500개_중간에 두번 걸림.xlsx'

df = pd.read_excel(file_name, header=None)
```

```python
sensor_list = []
time_list = []
singlejump_list = []
doublejump_list = []
looplist = []
for col in df:
    sensor_list.append(df[col][0])
    time_list.append(df[col][1])
    singlejump_list.append(0)
    looplist.append(0)
    doublejump_list.append(0)
```

```python
# How to detect single jumps
k = 0
j = 0
i = 0
# 바로 전에 싱글 점프가 진행되었는지
sinbool = False
while i < len(sensor_list)-4:
    if sensor_list[i:i+4] == [4,3,2,1]:
        if sinbool == False:
            if i == 0:
                k += 1
                singlejump_list[i:i+4] = [k,k,k,k]
                j = i+4
                i+=4
                sinbool = True
            # 에러처리 시간이 200을 넘어야한다
            elif sum(time_list[j:i]) > 200:
                k += 1
                singlejump_list[i:i+4] = [k,k,k,k]
                j = i+4
                i+=4
                sinbool = True
            else:
                i+=4
                sinbool = False
        else:
            k += 1
            singlejump_list[i:i+4] = [k,k,k,k]
            j = i+4
            i+=4
            sinbool = True
    else:
        i+=1
        sinbool = False
```

```python
print(singlejump_list)
```

```python
# How to detect double jumps
i = -1
loop = 0
dloop = 0
while i < len(sensor_list)-8:
    i+=1
    # 4,3,2,1을 두번 감지한다
    if sensor_list[i] == 4:
        # 같은 숫자가 여러번 나와도 된다
        while sensor_list[i] == 4:
            i += 1
        i -= 1
    else:
        continue

    a2 = i+1
    if sensor_list[a2] == 3:
        while sensor_list[a2] == 3:
            a2 += 1
        a2 -= 1
    else:
        continue

    a3 = a2+1
    if sensor_list[a3] == 2:
        while sensor_list[a3] == 2:
            a3 += 1
        a3-=1
    else:
        continue
    a4 = a3+1
    if sensor_list[a4] == 1:
        while sensor_list[a4] == 1:
            a4 += 1
        a4-=1
    else:
        continue
    a1 = a4 + 1
  
    if sensor_list[a1] == 4:
        while sensor_list[a1] == 4:
            a1 += 1
        a1 -= 1
    else:
        continue

    a2 = a1+1
    if sensor_list[a2] == 3:
        while sensor_list[a2] == 3:
            a2 += 1
        a2 -= 1
    else:
        continue

    a3 = a2+1
    if sensor_list[a3] == 2:
        while sensor_list[a3] == 2:
            a3 += 1
        a3-=1
    else:
        continue
    a4 = a3+1
    if sensor_list[a4] == 1:
        while sensor_list[a4] == 1:
            a4 += 1
        a4-=1
    else:
        continue
    
    loop +=1
    loop_list = []
    for _ in range(a4+1-i):
        loop_list.append(loop)
    looplist[i:a4+1] = loop_list
    # 더블 점프 하는 동안 시간이 800을 넘으면 안된다
    # 바로 전에 4,3,2,1 두 번이 감지 되어야 한다.
    if sum(time_list[i:a4+1]) < 800 and looplist[i-1] != 0:
        dloop += 1
        dloop_list = []
        for _ in range(a4+1-i):
            dloop_list.append(dloop)
        doublejump_list[i:a4+1] = dloop_list
    i = a4-1
```

```python
print(doublejump_list)
```

```python
for a,b in zip(singlejump_list, doublejump_list):
    print(a,b)
```
